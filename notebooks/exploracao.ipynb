```python
# Informações Básicas
print(f"Total de linhas carregadas: {len(influd25)}")
print(f"Total de colunas: {len(influd25.columns)}")
```

```python
# Colunas
for i, col in enumerate(influd25.columns, 1):
    print(f"{i:3}. {col}")
```

```python
# Colunas Datas
colunas_data = [col for col in influd25.columns if 'DT_' in col or 'DATA' in col.upper() or 'DOSE' in col]
for col in colunas_data:
    exemplo = influd25[col].dropna().head(5).tolist()
    print(f"  {col}: {exemplo}")
```

```python
# Colunas que achei importante olhando o dicionário de dados
colunas_importantes = {
    'DT_NOTIFIC': 'Data de notificação (para casos diários/mensais)',
    'EVOLUCAO': 'Evolução do caso - 1:Cura, 2:Óbito SRAG, 3:Óbito outras causas',
    'UTI': 'Se foi internado em UTI',
    'VACINA_COV': 'Se tomou vacina COVID',
    'VACINA': 'Se tomou vacina Gripe',
    'HOSPITAL': 'Se foi hospitalizado',
}

for col, desc in colunas_importantes.items():
    if col in influd25.columns:
        valores = influd25[col].value_counts().head(5)
        print(f"\n  {col} - {desc}")
        print(f"  Valores únicos: {influd25[col].nunique()}")
        print(f"  Valores mais frequentes:")
        for val, count in valores.items():
            print(f"    {val}: {count} ({count/len(influd25)*100:.1f}%)")
    else:
        print(f"\n  {col} - NÃO ENCONTRADA")
```

```python
# Dados Faltantes
missing = influd25.isnull().sum().sort_values(ascending=False).head(20)
for col, count in missing.items():
    pct = count / len(influd25) * 100
    print(f"  {col}: {count} ({pct:.1f}%)")
```

```python
# Periodo dos Dados
if 'DT_NOTIFIC' in influd25.columns:
    influd25['DT_NOTIFIC'] = pd.to_datetime(influd25['DT_NOTIFIC'], errors='coerce')
    print(f"  Data mais antiga: {influd25['DT_NOTIFIC'].min()}")
    print(f"  Data mais recente: {influd25['DT_NOTIFIC'].max()}")

    # Casos por mês
    print("\n  Casos por mês:")
    casos_mes = influd25.groupby(influd25['DT_NOTIFIC'].dt.to_period('M')).size().tail(12)
    for periodo, count in casos_mes.items():
        print(f"    {periodo}: {count} casos")
```
